```python
df.keys(),sunlight.keys(),aqi.keys()
```

```python
df["发病日期"] =  pd.to_datetime(df["发病日期"])
sunlight['时间'] = pd.to_datetime(sunlight['时间'])
aqi['统计日期'] = pd.to_datetime(aqi['统计日期'])
```

```python
# Merging the two dataframes on the conditions
merged_df = pd.merge(df, sunlight, left_on=['发病日期', '现在住址地市'], right_on=['时间', 'city'], how='left')

# Renaming '数值' to 'aqi'
merged_df = merged_df.rename(columns={'数值': 'sunlight'})

# Display the merged dataframe
merged_df
```

```python
# Merging the two dataframes on the conditions
merged_df = pd.merge(merged_df, aqi, left_on=['发病日期', '现在住址地市'], right_on=['统计日期', '城市名称'], how='left')

# Dropping unwanted columns
merged_df = merged_df.drop(columns=['统计日期', '城市代码', '时间', 'city', '城市名称', '空气质量级别', '首要污染物'])

# Display the merged dataframe
merged_df
```
